# Outliers Detection

```python
boston_describe=boston.describe().T
```

```python
boston_describe['Interquartile Range'] = 1.5*(boston_describe['75%'] - boston_describe['25%'])
```

```python
boston_describe['Major Outlier'] = (boston_describe['75%'] + boston_describe['Interquartile Range'])
boston_describe['Minor Outlier'] = (boston_describe['25%'] - boston_describe['Interquartile Range'])
```

```python
def remove_outlier(boston,variable):
    major_o = boston_describe.loc[variable,'Major Outlier']
    minor_o = boston_describe.loc[variable,'Minor Outlier']
    boston = boston.drop(boston[(boston[variable]>major_o) | (boston[variable]<minor_o)].index)
    return boston
```

```python
column_list = boston[['crim','zn','indus','rm','age','dis','tax','ptratio','lstat']]
```

```python
for column_name in column_list:
    boston = remove_outlier(boston, column_name)
```

# Prediction Algorithms

```python
boston = pd.read_csv("boston_loaded_1.csv")
```

```python
boston.head()
```

```python
# Creating Metrics Dataframe
metrics_boston = pd.DataFrame(index = ['RSquared_train', 'RSquared_test', 'RMS_train', 'RMS_test', 'MAE_train', 'MAE_test', 'MAPE_train', 'MAPE_test'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Splitting into training and testing dataset
X = boston.drop(['medv'],axis=1)
y = boston['medv']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# Function to print metrics
def print_training_metrics(**kwargs):
    for name, value in kwargs.items():
        value = float("{0:.2f}".format(value))
        print('{0} for Training Dataset is: {1}'.format(name, value))
def print_testing_metrics(**kwargs):
    for name, value in kwargs.items():
        value = float("{0:.2f}".format(value))
        print('{0} for Testing Dataset is: {1}'.format(name, value))
```

```python
# Function to print and add metrics to dataframe
def print_metrics(boston, model, r2_train, rms_train, mae_train, mape_train, r2_test, rms_test, mae_test, mape_test):
    boston[model] = [float("{0:.2f}".format(r2_train)), float("{0:.2f}".format(r2_test)),
                 float("{0:.2f}".format(rms_train)), float("{0:.2f}".format(rms_test)),
                 float("{0:.2f}".format(mae_train)), float("{0:.2f}".format(mae_test)),
                 float("{0:.2f}".format(mape_train)), float("{0:.2f}".format(mape_test))]
    return boston
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from math import sqrt
from sklearn.metrics import mean_absolute_error
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)

# Predicting and Calculating the Metrices for Prediction of Testing Dataset
prediction_test_lr = lr.predict(X_test)

r2_test_lr = r2_score(y_test, prediction_test_lr)
rms_test_lr = sqrt(mean_squared_error(y_test, prediction_test_lr))
mae_test_lr = mean_absolute_error(y_test,prediction_test_lr)
mape_test_lr = np.mean(np.abs((y_test - prediction_test_lr) / y_test)) * 100

# Predicting and Calculating the Metrices for Prediction of Training Dataset
prediction_train_lr = lr.predict(X_train)

r2_train_lr = r2_score(y_train, prediction_train_lr)
rms_train_lr = sqrt(mean_squared_error(y_train, prediction_train_lr))
mae_train_lr = mean_absolute_error(y_train,prediction_train_lr)
mape_train_lr = np.mean(np.abs((y_train - prediction_train_lr) / y_train)) * 100

# Printing the training and testing metrices
metrics_boston = print_metrics(metrics_boston, 'LR_Model', r2_train_lr, rms_train_lr, mae_train_lr, mape_train_lr, r2_test_lr, rms_test_lr, mae_test_lr, mape_test_lr)
metrics_boston
```

# Random forest Model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators = 500)
rf.fit(X_train, y_train)

# Predicting and Calculating the Metrices for Prediction of Testing Dataset
prediction_test_rf = rf.predict(X_test)

r2_test_rf = r2_score(y_test, prediction_test_rf)
rms_test_rf = sqrt(mean_squared_error(y_test, prediction_test_rf))
mae_test_rf = mean_absolute_error(y_test,prediction_test_rf)
mape_test_rf = np.mean(np.abs((y_test - prediction_test_rf) / y_test)) * 100

# Predicting and Calculating the Metrices for Prediction of Training Dataset
prediction_train_rf = rf.predict(X_train)

r2_train_rf = r2_score(y_train, prediction_train_rf)
rms_train_rf = sqrt(mean_squared_error(y_train, prediction_train_rf))
mae_train_rf = mean_absolute_error(y_train,prediction_train_rf)
mape_train_rf = np.mean(np.abs((y_train - prediction_train_rf) / y_train)) * 100

# Printing the training and testing metrices
print('Random Forest Model\n')
metrics_boston = print_metrics(metrics_boston, 'RF_Model', r2_train_rf, rms_train_rf, mae_train_rf, mape_train_rf, r2_test_rf, rms_test_rf, mae_test_rf, mape_test_rf)
metrics_boston
```

## Recursive Feature Elimination
> [Source](https://machinelearningmastery.com/feature-selection-in-python-with-scikit-learn/):
>The Recursive Feature Elimination (RFE) method is a feature selection approach. It works by recursively removing attributes and building a model on those attributes that remain. It uses the model accuracy to identify which attributes (and combination of attributes) contribute the most to predicting the target attribute.

>This recipe shows the use of RFE to select 3 attributes.

```python
# Creating Features Dataframe
all_feature = X.columns.tolist()
feature_boston = pd.DataFrame(index = all_feature)
```

```python
# Recursive Feature Elimination
from sklearn.feature_selection import RFE
from sklearn.linear_model import (LinearRegression, Ridge, Lasso)
from sklearn.ensemble import RandomForestRegressor
```

```python
# Create a base classifier used to evaluate a subset of attributes
model_lr = LinearRegression()
model_lasso = Lasso()
model_rf = RandomForestRegressor(n_estimators=50)


# Create the RFE model and select 3 attributes
rfe_lr = RFE(model_lr, n_features_to_select = 3)
rfe_lr = rfe_lr.fit(X, y)

rfe_lasso = RFE(model_lasso, n_features_to_select = 3)
rfe_lasso = rfe_lasso.fit(X, y)

rfe_rf = RFE(model_rf, n_features_to_select = 3)
rfe_rf = rfe_rf.fit(X, y)

# Summarize the selection of the attributes
feature_boston['RFE_LR_Rank'] = rfe_lr.ranking_
feature_boston['RFE_Lasso_Rank'] = rfe_lasso.ranking_
feature_boston['RFE_RF_Rank'] = rfe_rf.ranking_

feature_boston
```

# Contrast Feature Selection

# Using Boruta Library

Boruta is an all relevant feature selection method, while most other are minimal optimal; this means it tries to find all features carrying information usable for prediction, rather than finding a possibly compact subset of features on which some classifier has a minimal error.

```python
!pip install boruta
```

```python
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
```

```python
# Change X and y to its values
X_boruta = X.values
y_boruta = y.values

# Define random forest classifier, with utilising all cores and sampling in proportion to y labels
rfc = RandomForestClassifier(n_jobs=-1)
 
# Define Boruta feature selection method
feature_selector = BorutaPy(rfc, n_estimators='Auto', verbose=2)
 
# Find all relevant features
feature_selector.fit(X_boruta, y_boruta)
 
# Check ranking of features
feature_boston['Boruta_Rank'] = feature_selector.ranking_
feature_boston
```
